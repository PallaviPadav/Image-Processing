# Object identification from the image

jpeg is taken as input image and the object present in it is identified using the box.

```python
import pixellib
from pixellib.instance import instance_segmentation

segment_image = instance_segmentation()
segment_image.load_model("mask_rcnn_coco.h5")
segment_image.segmentImage('sydney1.jpg', output_image_name = 'sydney1_op.jpg', show_bboxes = True)
```

```python
segment_image.segmentImage('singapore1.jpg', output_image_name = 'singapore1_op.jpg', show_bboxes = True)
```

```python
segment_image.segmentImage('testing.jpg', output_image_name = 'testing_op.jpg', show_bboxes = True)
```

```python
results, output = segment_image.segmentImage('sydney.jpg', output_image_name = 'sydney_op.jpg', show_bboxes = True)
print(results)
```
